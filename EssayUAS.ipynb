```python
print(df.head())

print("Rata-rata math score.",df['math score'].mean())
print("Standar deviasi math score.",df['math score'].std())
print("Median math score",df['math score'].median())
print("Nilai minimum math score",df['math score'].min())
print("Nilai max math score",df['math score'].max())

print("jumlah data: ", df['gender'].count())
```

```python
from sklearn.linear_model import LinearRegression
df = pd.read_csv('StudentsPerformance.csv')
df

X = df[['reading score']]
y = df[['writing score']]

model = LinearRegression()
model.fit(X,y)

plt.scatter(X,y, color='blue', marker='o', label='Data Asli')
plt.plot(X, model.predict(X), color='red', label='Garis Regresi')
plt.xlabel('reading score')
plt.ylabel('writing score')
plt.title('Regresi Linier: Reading Vs Writing Score')
plt.grid(True)
plt.legend
plt.show
```
